```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model
```

```python
#read the features and labels
train_features = pd.read_csv('processed_train_features.csv')
test_features = pd.read_csv('processed_test_features.csv')
train_labels = pd.read_csv('train_labels.csv')["homo_lumo_gap"].to_numpy()
ids = pd.read_csv('test_features.csv')['Id'].to_numpy()

train_features = train_features.to_numpy()
test_features = test_features.to_numpy()
```

```python
#initilaize callback for early stopping
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=15)

#training test split
X_train, X_test, y_train, y_test = train_test_split(train_features, train_labels, test_size=0.2, random_state = 42)

#desing a model
final_model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(1, activation = 'linear')
])

#compile a model
final_model.compile(loss='mse',optimizer=tf.keras.optimizers.Adam())

#train the model
history = final_model.fit(train_features, train_labels, epochs= 2000,
                         validation_data=(X_test, y_test), callbacks=[callback])
```

```python
#visualize the performance of the model
def plot_loss(history):
    
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    #plt.plot(history.history['rmse'], label='rmse')
    #plt.plot(history.history['val_rmse'], label='val_rmse')
    plt.ylim([0,0.2])
    plt.xlabel('Epoch')
    plt.ylabel('Error [MPG]')
    plt.legend()
    plt.grid(True)


plot_loss(history)

def evaluation(model, X_test, y_test):
    prediction = model.predict(X_test)
    mae = mean_absolute_error(y_test, prediction)
    mse = mean_squared_error(y_test, prediction)
    
    plt.figure(figsize=(15, 10))
    plt.plot(prediction[:100], "red", label="prediction", linewidth=2.0)
    plt.plot(y_test[:100], 'green', label="actual", linewidth=2.0)
    plt.legend()
    plt.ylabel('logP')
    plt.title("MAE {}, MSE {}".format(round(mae, 4), round(mse, 4)))
    plt.show()
    
    print('MAE score:', round(mae, 4))
    print('MSE score:', round(mse,4))
    
evaluation(final_model, X_train, y_train)
evaluation(final_model, X_test, y_test)
```

```python
#calculate and store the final prediction
predictions = final_model.predict(test_features)
predictions = predictions.reshape(-1)
df_handin = pd.DataFrame({'Id': ids, 'y': predictions})
df_handin.to_csv('predictions.csv',index = False)
print(predictions)
```
